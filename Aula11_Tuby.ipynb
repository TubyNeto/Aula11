```python
#def GaussJacobi

import numpy as np
import matplotlib.pyplot as plt

def GaussJacobi(A, b, n):
  xo = np.zeros(n)
  xk = np.zeros(n)
  continueIter = True
  counter = 0

  xk = np.copy(xo)
  xk1 = np.zeros(n)
  
  #Continuamos as iterações para o cálculo da solução até que a condição da diferença mínima seja satisfeita nas aproximações
  while (continueIter):
    counter+=1
    for i in range (n):
      xk1[i] = b[i]/A[i, i]
      for j in range (n):
        if (i != j):
          xk1[i] = xk1[i] - ((A[i, j] * xk[j])/A[i, i])  
     

    print("Iteration", counter, xk) 

    if (np.allclose(xk1, xk, 0.35, 0.)): 
        xk = np.copy(xk1)
        break
    else: xk = np.copy(xk1)

  print('Solution = \n' , xk)


A  = np.matrix([[5, 2, 2], [1, 3, 1], [0, 6, 8] ])
b  = np.array([3, -2, -6])
n = np.shape(A)[0]

GaussJacobi(A, b, n)
```

```python
#def GaussSeidel

def GaussSeidel(A, b, n):
  xo = np.zeros(n)
  xk = np.zeros(n)
  continueIter = True
  counter = 0

  xk = np.copy(xo)
  xk1 = np.zeros(n)

  #Continuamos as iterações para o cálculo da solução até que a condição da diferença mínima seja satisfeita nas aproximações
  while (continueIter):
    counter+=1
    for i in range (n):
      xk1[i] = b[i]/A[i, i]
      for j in range (n):
        if (i > j):
            xk1[i] = xk1[i] - ((A[i, j] * xk1[j])/A[i, i])
        elif (i != j):
          xk1[i] = xk1[i] - ((A[i, j] * xk[j])/A[i, i])  
     

    print("Iteration", counter, xk) 

    if (np.allclose(xk1, xk, 0.35, 0.)): 
        xk = np.copy(xk1)
        break
    else: xk = np.copy(xk1)

  print('Solution = \n' , xk1)


A  = np.matrix([[5, 2, 2], [1, 3, 1], [0, 6, 8] ])
b  = np.array([3, -2, -6])
n = np.shape(A)[0]

GaussSeidel(A, b, n)
```

```python
#def jacobiMethod
#Utilizamos as matrizes C e g

def jacobiMethod(A, b, n):
  e = 0.05
  xo = np.zeros(n)
  xk = np.zeros(n)
  C = np.zeros((n, n))
  g = np.zeros(n)
  continueIter = True
  counter = 0

  #Calcula a matriz C
  for i in range (n):
    for j in range (n):
      if (i==j): C[i, j] = 0
      else: C[i, j] = (A[i, j]/A[i, i])*(-1)

  #Calcula a matriz g
  for i in range (n):
    g[i] = b[i]/A[i, i]

  xk = np.copy(xo)
  xk1 = np.copy(xo)
  diff = np.zeros(n)
  
  #Continuamos as iterações para o cálculo da solução até que a condição da diferença mínima seja satisfeita nas aproximações
  while (continueIter):
    counter+=1
    xk1 = np.dot(C, xk) + g
    for i in range (n):
      diff[i] = xk1[i] - xk[i]

    max_diff = np.amax(abs(diff))
    max_xk1 = np.amax(abs(xk1))

    if (max_diff/max_xk1 < e): 
        xk = np.copy(xk1)
        break
    else: xk = np.copy(xk1)

  print('Solution = \n' , xk)


A  = np.matrix([[5, 2, 2], [1, 3, 1], [0, 6, 8] ])
b  = np.array([3, -2, -6])
n = np.shape(A)[0]
jacobiMethod(A, b, n)
```
